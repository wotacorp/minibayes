# Distribution Gallery

This notebook showcases all probability distributions available in minibayes.

**Total: 13 distributions**
- 11 continuous (REAL, POSITIVE, UNIT, BOUNDED)
- 2 discrete (BINARY, NATURAL)

```python
import numpy as np
import matplotlib.pyplot as plt
from minibayes import dist, viz
```

---

## 1. Continuous - Unbounded (REAL Support)

These distributions have support on the entire real line $(-\infty, +\infty)$.

### 1.1 Normal Distribution

**Parameters:** `loc` (mean), `scale` (std dev)  
**Use case:** Default prior for unbounded parameters

```python
distributions = {
    "Normal(0, 0.5)": dist.Normal(0, 0.5),
    "Normal(0, 1)": dist.Normal(0, 1),
    "Normal(0, 2)": dist.Normal(0, 2),
    "Normal(1, 1)": dist.Normal(1, 1),
}
viz.plot_distribution(distributions)
plt.title("Normal Distribution")
plt.show()
```

### 1.2 Student-t Distribution

**Parameters:** `df` (degrees of freedom), `loc`, `scale`  
**Use case:** Robust regression with heavy tails (outliers)

```python
distributions = {
    "StudentT(df=1) [Cauchy]": dist.StudentT(1, 0, 1),
    "StudentT(df=3)": dist.StudentT(3, 0, 1),
    "StudentT(df=10)": dist.StudentT(10, 0, 1),
    "Normal(0, 1)": dist.Normal(0, 1),
}
viz.plot_distribution(distributions)
plt.title("Student-t Distribution (heavier tails with lower df)")
plt.show()
```

### 1.3 Cauchy Distribution

**Parameters:** `loc` (median), `scale`  
**Use case:** Very robust priors, pathological heavy tails

**Note:** Mean and variance are undefined!

```python
distributions = {
    "Cauchy(0, 0.5)": dist.Cauchy(0, 0.5),
    "Cauchy(0, 1)": dist.Cauchy(0, 1),
    "Cauchy(0, 2)": dist.Cauchy(0, 2),
}
x = np.linspace(-10, 10, 200)
viz.plot_distribution(distributions, x=x)
plt.title("Cauchy Distribution (undefined mean/variance)")
plt.show()
```

### 1.4 Laplace Distribution

**Parameters:** `loc` (mean), `scale`  
**Use case:** Sparse priors, L1 regularization (Lasso)

```python
distributions = {
    "Laplace(0, 0.5)": dist.Laplace(0, 0.5),
    "Laplace(0, 1)": dist.Laplace(0, 1),
    "Laplace(0, 2)": dist.Laplace(0, 2),
    "Normal(0, 1)": dist.Normal(0, 1),
}
viz.plot_distribution(distributions)
plt.title("Laplace vs Normal (sharper peak, heavier tails)")
plt.show()
```

## 2. Continuous - Positive (POSITIVE Support)

These distributions have support on $(0, +\infty)$. Automatically use LogTransform.

### 2.1 HalfNormal Distribution

**Parameters:** `scale`  
**Use case:** Weakly informative prior for scale/std parameters

```python
distributions = {
    "HalfNormal(0.5)": dist.HalfNormal(0.5),
    "HalfNormal(1)": dist.HalfNormal(1),
    "HalfNormal(2)": dist.HalfNormal(2),
}
viz.plot_distribution(distributions)
plt.title("HalfNormal Distribution")
plt.show()
```

### 2.2 Exponential Distribution

**Parameters:** `rate` (λ)  
**Use case:** Waiting times, memoryless processes, sparse priors

```python
distributions = {
    "Exponential(rate=0.5)": dist.Exponential(0.5),
    "Exponential(rate=1)": dist.Exponential(1),
    "Exponential(rate=2)": dist.Exponential(2),
}
viz.plot_distribution(distributions)
plt.title("Exponential Distribution")
plt.show()
```

### 2.3 Gamma Distribution

**Parameters:** `shape` (α), `rate` (β)  
**Use case:** Precision parameters, positive quantities with flexible shape

```python
distributions = {
    "Gamma(1, 1) [Exponential]": dist.Gamma(1, 1),
    "Gamma(2, 1)": dist.Gamma(2, 1),
    "Gamma(3, 1)": dist.Gamma(3, 1),
    "Gamma(5, 1)": dist.Gamma(5, 1),
}
x = np.linspace(0.01, 8, 200)
viz.plot_distribution(distributions, x=x)
plt.title("Gamma Distribution (shape controls skewness)")
plt.show()
```

### 2.4 LogNormal Distribution

**Parameters:** `loc` (μ of log), `scale` (σ of log)  
**Use case:** Prices, concentrations, multiplicative processes

```python
distributions = {
    "LogNormal(0, 0.25)": dist.LogNormal(0, 0.25),
    "LogNormal(0, 0.5)": dist.LogNormal(0, 0.5),
    "LogNormal(0, 1)": dist.LogNormal(0, 1),
}
x = np.linspace(0.01, 6, 200)
viz.plot_distribution(distributions, x=x)
plt.title("LogNormal Distribution (skewed positive)")
plt.show()
```

### 2.5 InverseGamma Distribution

**Parameters:** `shape` (α), `scale` (β)  
**Use case:** Variance priors (conjugate for Normal variance)

```python
distributions = {
    "InverseGamma(2, 1)": dist.InverseGamma(2, 1),
    "InverseGamma(3, 1)": dist.InverseGamma(3, 1),
    "InverseGamma(5, 1)": dist.InverseGamma(5, 1),
}
viz.plot_distribution(distributions)
plt.title("InverseGamma Distribution")
plt.show()
```

## 3. Continuous - Unit Interval (UNIT Support)

These distributions have support on $(0, 1)$. Automatically use LogitTransform.

### 3.1 Beta Distribution

**Parameters:** `alpha` (α), `beta` (β)  
**Use case:** Probabilities, proportions, rates between 0 and 1

```python
distributions = {
    "Beta(1, 1) [Uniform]": dist.Beta(1, 1),
    "Beta(2, 2) [Symmetric]": dist.Beta(2, 2),
    "Beta(5, 2) [Right skew]": dist.Beta(5, 2),
    "Beta(2, 5) [Left skew]": dist.Beta(2, 5),
    "Beta(0.5, 0.5) [U-shape]": dist.Beta(0.5, 0.5),
}
viz.plot_distribution(distributions)
plt.title("Beta Distribution (flexible shape on [0,1])")
plt.show()
```

---

## 4. Continuous - Bounded (BOUNDED Support)

These distributions have support on a bounded interval $[a, b]$. Use AffineTransform.

### 4.1 Uniform Distribution

**Parameters:** `low`, `high`  
**Use case:** Non-informative priors on bounded parameters

```python
# Note: Uniform distributions with different bounds need custom x ranges
fig, ax = plt.subplots(figsize=(8, 4))

uniforms = [
    ("Uniform(0, 1)", dist.Uniform(0, 1)),
    ("Uniform(0, 2)", dist.Uniform(0, 2)),
    ("Uniform(-1, 1)", dist.Uniform(-1, 1)),
]

for label, d in uniforms:
    viz.plot_distribution({label: d}, ax=ax)

plt.title("Uniform Distribution")
plt.xlim(-2, 3)
plt.show()
```

---

## 5. Discrete Distributions

These distributions are primarily used in **likelihood functions** for discrete data.

`viz.plot_distribution()` automatically detects discrete distributions and uses bar plots.

### 5.1 Bernoulli Distribution

**Parameters:** `prob` (p)  
**Use case:** Binary outcomes, classification likelihoods

```python
distributions = {
    "p=0.2": dist.Bernoulli(0.2),
    "p=0.5": dist.Bernoulli(0.5),
    "p=0.8": dist.Bernoulli(0.8),
}
viz.plot_distribution(distributions)
plt.title("Bernoulli Distribution")
plt.show()
```

### 5.2 Poisson Distribution

**Parameters:** `rate` (λ)  
**Use case:** Count data, rare events

```python
distributions = {
    "Poisson(λ=1)": dist.Poisson(1),
    "Poisson(λ=3)": dist.Poisson(3),
    "Poisson(λ=5)": dist.Poisson(5),
    "Poisson(λ=10)": dist.Poisson(10),
}
viz.plot_distribution(distributions, k_max=20)
plt.title("Poisson Distribution")
plt.show()
```

```python
reference = """
| Distribution    | Support      | Parameters              | Transform       | Use Case                          |
|-----------------|--------------|-------------------------|-----------------|-----------------------------------|
| Normal          | (-∞, +∞)     | loc, scale              | Identity        | Unbounded parameters              |
| StudentT        | (-∞, +∞)     | df, loc, scale          | Identity        | Robust regression, heavy tails    |
| Cauchy          | (-∞, +∞)     | loc, scale              | Identity        | Very robust priors                |
| Laplace         | (-∞, +∞)     | loc, scale              | Identity        | Sparse priors (L1)                |
| HalfNormal      | (0, +∞)      | scale                   | Log             | Scale parameters                  |
| Exponential     | (0, +∞)      | rate                    | Log             | Waiting times, sparse             |
| Gamma           | (0, +∞)      | shape, rate             | Log             | Precision, positive quantities    |
| LogNormal       | (0, +∞)      | loc, scale              | Log             | Prices, concentrations            |
| InverseGamma    | (0, +∞)      | shape, scale            | Log             | Variance priors                   |
| Beta            | (0, 1)       | alpha, beta             | Logit           | Probabilities, proportions        |
| Uniform         | [low, high]  | low, high               | Affine          | Non-informative, bounded          |
| Bernoulli       | {0, 1}       | prob                    | Identity*       | Binary outcomes (likelihood)      |
| Poisson         | {0, 1, ...}  | rate                    | Identity*       | Count data (likelihood)           |

*Discrete distributions are typically used in likelihoods, not as sampled parameters.
"""
print(reference)
```

## 7. Sampling Example

```python
rng = np.random.default_rng(42)

# Sample from each distribution
samples = {
    "Normal(0, 1)": dist.Normal(0, 1).sample(1000, rng),
    "StudentT(3, 0, 1)": dist.StudentT(3, 0, 1).sample(1000, rng),
    "Gamma(2, 1)": dist.Gamma(2, 1).sample(1000, rng),
    "Beta(2, 5)": dist.Beta(2, 5).sample(1000, rng),
}

fig, axes = plt.subplots(2, 2, figsize=(10, 6))
for ax, (name, s) in zip(axes.flat, samples.items()):
    ax.hist(s, bins=50, density=True, alpha=0.7)
    ax.set_title(name)
    ax.set_ylabel("Density")

plt.tight_layout()
plt.show()
```
